```python
print('tipe financial statement')
pd.DataFrame(pd.unique(data['type']))
```

```python
# buat kolom baru, gabungan antara account dan type
data['account_type'] = data['account']+'_'+data['type']
data.head()
```

### reshaping data

```python
# ambil data 2021 saja
data2021 = data[['symbol', 'account_type', '2021-12-31']].copy()
# reshaping data, kolom" data yang baru merupakan value dari 'type' dari df yang lama
dataReshaped = data2021.pivot_table(
    index='symbol',
    columns='account_type',
    values='2021-12-31'
).reset_index()
dataReshaped.head()
```
